```python
#changeing the coulmn name in df1 to match df2
df1 = df1.rename(columns={
    'Age': 'Age',
    'Gender': 'Gender ',
    'Heart Rate':'Heart rate',
    'Chest Pain Type':'Chest pain type',
    'Blood Pressure':'Blood pressure',
     'Blood Sugar':' Blood sugar',
    'Heart Disease':'Result',
    'Cholesterol':'Cholesterol',

})
df1.columns = df1.columns.str.strip().str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.strip().str.lower().str.replace(' ', '_')

# cheaking consistant coulmns naming
print(df1.columns.tolist())
print(df2.columns.tolist())
```

# **Data Preprocessing and Cleaning for df1**

```python
# Select numeric columns
numeric_cols = df1.select_dtypes(include=['int64', 'float64']).columns.tolist()

#  Detect and handle outliers
for col in numeric_cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df1[(df1[col] < lower_bound) | (df1[col] > upper_bound)]
    print(f"Column: {col}")
    print(f"Number of outliers: {len(outliers)}")
    print(f"Outlier values:\n{outliers[col].values}\n")

    df1[col] = df1[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
duplicate_count = df1.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Display the duplicate rows if any exist
if duplicate_count > 0:
    print("\nDuplicate rows:")
    display(df1[df1.duplicated(keep=False)])
else:
    print(" No duplicate rows found.")
```

```python
# Identify numeric and categorical columns

numeric_cols = df1.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_cols = df1.select_dtypes(include='object').columns.tolist()

# Separate target column for safe transformation
target_col = 'result'
if target_col in numeric_cols:
    numeric_cols.remove(target_col)
if target_col in categorical_cols:
    categorical_cols.remove(target_col)


# Force numeric columns to float
for col in numeric_cols:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')

#  Numeric imputation with mean becaues no outliers are detected
num_imputer = SimpleImputer(strategy='mean')
df1[numeric_cols] = num_imputer.fit_transform(df1[numeric_cols])
df1[numeric_cols] = df1[numeric_cols].round(2)

# Categorical imputation with mode
cat_imputer = SimpleImputer(strategy='most_frequent')
df1[categorical_cols] = cat_imputer.fit_transform(df1[categorical_cols])

#  Impute target column with mode and convert to int
df1[[target_col]] = cat_imputer.fit_transform(df1[[target_col]])
df1[target_col] = df1[target_col].astype(int)
```

# **Data Preprocessing and Cleaning for df2**

```python
numeric_cols = df2.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df2.select_dtypes(include='object').columns.tolist()
if 'result' in numeric_cols:
    numeric_cols.remove('result')
```

```python
# Identify numeric columns
numeric_cols = df2.select_dtypes(include=['int64', 'float64']).columns.tolist()

#  Numeric imputation (median)
for col in numeric_cols:
    df2[col] = df2[col].fillna(df2[col].median())

# Handle outliers (clip using IQR)

for col in numeric_cols:
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df2[col] = df2[col].clip(lower, upper)

# Round numeric columns to 2 decimals

df2[numeric_cols] = df2[numeric_cols].round(2)
print("Missing values after preprocessing:\n ",df2[numeric_cols].isnull().sum())
df2['gender'] = df2['gender'].astype(int)
```

```python
# Count total number of duplicate rows in the dataset
duplicate_count = df2.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Display the duplicate rows if any exist
if duplicate_count > 0:
    print("\nDuplicate rows:")
    display(df2[df2.duplicated(keep=False)])
else:
    print("No duplicate rows found.")
```
